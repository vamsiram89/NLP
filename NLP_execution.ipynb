```python
text = "beckgammon is one of the oldest know board game. its history can be traced back nearly 500 years to archelogical discovers in the middle east. it is a two player game where each player has fifteen checkers which moves between twenty-four points according to the role two dice."
sentence = nltk.sent_tokenize(text)
for s in sentence:
  print(s)
  print()
```

### word tokenize

```python
for w in sentence:
  words = nltk.word_tokenize(w)
  print(words)
  print()
```

### stemming

Lemmatization

```python
from nltk.corpus import stopwords
import string
```

```python
# stemming is crude approach where it wont works always because it remove last letter like, ing,ed,er,es
#mostly steming won't works with simple past tense to avide this problem we use
#lemmatization it is a rule based approach
```

```python
from nltk.stem import PorterStemmer,WordNetLemmatizer
stop_words = set(stopwords.words('english'))
print(stop_words)

lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()
for w in sentence:
  words = nltk.word_tokenize(w)
  # remove the punctuations
  words = [word for word in words if word not in string.punctuation]
  # remove the stop words
  words = [word for word in words if word.lower not in stop_words]
  # get the root words from the above words
  lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
  stemmed_words = [stemmer.stem(word) for word in words]
  print("original words:",w)
  print("lemmatized:-",lemmatized_words)
  print("stemming:-",stemmed_words)
  print()

```

**bag of words** ---> is the way we used to convert text data into numerical forms for ML
--> it is used to count the text data in sentence
--> there are 2 ways
--> 1.set  vocabulary
--> 2.n-gram approach





```python
document = ['good movie good screeplay','good movie','bad movie sleep']
# import the library to convert the text into numbers
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd


# design the vocubulary
count_vectorizer = CountVectorizer(binary = False)

# create the bag of words
bag_of_words = count_vectorizer.fit_transform(document)

# show the bag of words with numerical forms
feature_names = count_vectorizer.get_feature_names_out()

df = pd.DataFrame(bag_of_words.toarray(),columns = feature_names)
print(df)
```
